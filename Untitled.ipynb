```python
def get_data(folder_path, first_item, last_item, prefix):
    imgs = []
    for i in range(first_item, last_item+1):
        imgs.append((cv.imread(folder_path + prefix + str(i) + '.jpg', 1), cv.imread(folder_path + prefix + str(i) + '.jpg', 0)))
        
    with open(folder_path + 'names.txt', encoding='utf8') as f:
        names = list(map(lambda x: x.rstrip(), f.readlines()))
    return imgs, names
```

```python
def prepareOpenFace(useCuda=False, gpuDevice=0, useMultiGPU=False):
    model = netOpenFace(useCuda, gpuDevice)
    model.load_state_dict(torch.load(os.path.join(containing_dir, 'openface.pth'), map_location='cpu'))

    if useMultiGPU:
        model = nn.DataParallel(model)

    return model
```

```python
class CombNN(nn.Module):
    def __init__(self, input_size, output_size):
        super().__init__()
        self.input_size = input_size
        self.output_size = output_size
        self.model = prepareOpenFace().eval()
        self.lin = nn.Linear(self.input_size, self.output_size)
    
    def forward(self, input_data, x2_batch):
        return F.softmax(self.lin(torch.cat([self.model(input_data)[0], x2_batch], dim=1)))
```

```python
def prep(img):
    img = cv.resize(img, (96, 96), interpolation=cv.INTER_LINEAR)
    img = numpy.transpose(img, (2, 0, 1))
    img = img.astype(numpy.float32) / 255.0
    I_ = torch.from_numpy(img).unsqueeze(0)
    return I_
```

```python
def find_face(image):
        fc = cv.CascadeClassifier('../haarcascade_frontalface_default.xml')
        gray = image[1]
        
        faces = fc.detectMultiScale(gray, 1.25, 5)
        x, y, w, h = faces[0]
        img = image[0][y:y+h, x:x+w]
        
        I_ = prep(img)
        return I_
```

```python
def imgs2faces(imgs):
    faces = []
    i=1
    for img in imgs:
        try:
            faces.append(find_face(img).numpy()[0])
        except:
            print("No face found! iteration ", i)
            faces.append(prep(img[0]).numpy()[0])
        i+=1
    return torch.tensor(np.stack(faces, axis=0))
```

```python
history = []

for i in range(1000):
    ix = np.random.randint(0, len(data1), 16)
    x_batch = torch.tensor(faces[ix], dtype=torch.float32)
    y_batch = torch.tensor(y[ix], dtype=torch.int64)
    x2_batch = torch.tensor(data[ix], dtype=torch.float32)
    
    y_batch_ohe = ohe.transform(y_batch.reshape(-1, 1))
    y_batch_ohe = torch.tensor(y_batch_ohe, dtype=torch.float32)
    y_predicted = comb(x_batch, x2_batch)
    crossentropy = -(torch.log(y_predicted)*y_batch_ohe)
    crossentropy = crossentropy.sum(dim=1)
    loss = torch.mean(crossentropy)
    
    loss.backward()
    opt.step()
    opt.zero_grad()
    
    history.append(loss.data.numpy())
    
    print("step #%i | mean loss = %.3f" % (i, np.mean(history[-10:])))
```
